```python
from datetime import datetime, timedelta
import pytz


# 使用者輸入（目前不影響計算，但保留）
minute_input = int(input("還有幾分鐘？"))
second_input = int(input("還有幾秒？"))

# 定義台北時區
taipei_tz = pytz.timezone('Asia/Taipei')
# 取得現在時間
now = datetime.now(taipei_tz)

# 加上 5 分鐘
mushroom_time = now + timedelta(minutes = minute_input+5,seconds=second_input)

# 格式化時間（例如 14:30:25）
formatted_time = mushroom_time.strftime("%H:%M:%S")

# 輸出結果
print(f"蘑菇會在 {formatted_time} 時出現")
```

```python
# 輸入座標
xPosition = float(input("Enter X position:  "))
yPosition = float(input("Enter Y position:  "))
zPosition = float(input("Enter Z position:  "))
# 定義基本資訊
numberOfCharge = int(input("Enter how many coulombs(integer)?  "))
portionsPerUnits = int(input("How many parts you want for each side?  "))

# 初始化常數
epsilon_0 = 8.854187817*10**-12
elementaryCharge = 1.602176634*10**-19
coulombConstant = 8.987551787*10**9
areaForEachPart = 1/(portionsPerUnits**2)
startPosition = 1/(2*portionsPerUnits)
totalFlux = 0.0

# 計算理論值
def theoryValue():
  if((xPosition<1.0 and xPosition>0)and(yPosition<1.0 and yPosition>0)and(zPosition<1.0 and zPosition>0)):
    theoryValue = numberOfCharge*elementaryCharge/epsilon_0
  else:
    theoryValue = 0.0
  return theoryValue

def findCosValue(dx,dy,dz,nx,ny,nz):
  # lengthOfNormalVector = 1
  lengthOfPositionVector = ((xPosition-dx)**2+(yPosition-dy)**2+(zPosition-dz)**2)**0.5
  productValue = ((dx-xPosition)*(nx)+(dy-yPosition)*(ny)+(dz-zPosition)*(nz))
  cosValue = productValue/lengthOfPositionVector
  return cosValue


def findElectricField(dx,dy,dz):
  lengthSquare = ((xPosition-dx)**2+(yPosition-dy)**2+(zPosition-dz)**2)**-1
  electricValue = coulombConstant*numberOfCharge*elementaryCharge*lengthSquare
  return electricValue

# downFace: z=0
# downFace初始化
downFace_X = []
downFace_Y = []
downFace_Z = []
# downFace定位
for xIndex in range(1,portionsPerUnits+1):
  for yIndex in range(1,portionsPerUnits+1):
    downFace_X.append(startPosition*((xIndex*2)-1))
    downFace_Y.append(startPosition*((yIndex*2)-1))
    downFace_Z.append(0)

# 計算結果，電場*面積*cosTheta = 單位電流
for i in range(0,(portionsPerUnits**2)):
  totalFlux = totalFlux + findElectricField(downFace_X[i],downFace_Y[i],downFace_Z[i])*areaForEachPart*findCosValue(downFace_X[i],downFace_Y[i],downFace_Z[i],0,0,-1)

# upFace: z=1

upFace_X = []
upFace_Y = []
upFace_Z = []

for xIndex in range(1,portionsPerUnits+1):
  for yIndex in range(1,portionsPerUnits+1):
    upFace_X.append(startPosition*((xIndex*2)-1))
    upFace_Y.append(startPosition*((yIndex*2)-1))
    upFace_Z.append(1)

for i in range(0,(portionsPerUnits**2)):
  totalFlux = totalFlux + findElectricField(upFace_X[i],upFace_Y[i],upFace_Z[i])*areaForEachPart*findCosValue(upFace_X[i],upFace_Y[i],upFace_Z[i],0,0,1)

# leftFace = x=0
leftFace_X = []
leftFace_Y = []
leftFace_Z = []

for yIndex in range(1,portionsPerUnits+1):
  for zIndex in range(1,portionsPerUnits+1):
    leftFace_X.append(0)
    leftFace_Y.append(startPosition*((yIndex*2)-1))
    leftFace_Z.append(startPosition*((zIndex*2)-1))

for i in range(0,(portionsPerUnits**2)):
  totalFlux = totalFlux + findElectricField(leftFace_X[i],leftFace_Y[i],leftFace_Z[i])*areaForEachPart*findCosValue(leftFace_X[i],leftFace_Y[i],leftFace_Z[i],-1,0,0)

# rightFace = x=1
rightFace_X = []
rightFace_Y = []
rightFace_Z = []

for yIndex in range(1,portionsPerUnits+1):
  for zIndex in range(1,portionsPerUnits+1):
    rightFace_X.append(1)
    rightFace_Y.append(startPosition*((yIndex*2)-1))
    rightFace_Z.append(startPosition*((zIndex*2)-1))

for i in range(0,(portionsPerUnits**2)):
  totalFlux = totalFlux + findElectricField(rightFace_X[i],rightFace_Y[i],rightFace_Z[i])*areaForEachPart*findCosValue(rightFace_X[i],rightFace_Y[i],rightFace_Z[i],1,0,0)

# backFace = y=0
backFace_X = []
backFace_Y = []
backFace_Z = []

for xIndex in range(1,portionsPerUnits+1):
  for zIndex in range(1,portionsPerUnits+1):
    backFace_X.append(startPosition*((xIndex*2)-1))
    backFace_Y.append(0)
    backFace_Z.append(startPosition*((zIndex*2)-1))

for i in range(0,(portionsPerUnits**2)):
  totalFlux = totalFlux + findElectricField(backFace_X[i],backFace_Y[i],backFace_Z[i])*areaForEachPart*findCosValue(backFace_X[i],backFace_Y[i],backFace_Z[i],0,-1,0)

# frontFace = y=1
frontFace_X = []
frontFace_Y = []
frontFace_Z = []

for xIndex in range(1,portionsPerUnits+1):
  for zIndex in range(1,portionsPerUnits+1):
    frontFace_X.append(startPosition*((xIndex*2)-1))
    frontFace_Y.append(1)
    frontFace_Z.append(startPosition*((zIndex*2)-1))

for i in range(0,(portionsPerUnits**2)):
  totalFlux = totalFlux + findElectricField(frontFace_X[i],frontFace_Y[i],frontFace_Z[i])*areaForEachPart*findCosValue(frontFace_X[i],frontFace_Y[i],frontFace_Z[i],0,1,0)

print("--------------------------------------------------------------------")
print("Analysed data here:")
print(f"The computed value is: {totalFlux}")
print(f"The theory number is:  {theoryValue()}")
print(f"The difference is: {totalFlux-theoryValue()}")
print("--------------------------------------------------------------------")
```

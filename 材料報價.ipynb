```python
def getPrice():
    x = float(input("多長？(單位是毫米喔！)"))
    y = float(input("多寬？(單位是毫米喔！)"))
    z = float(input("多高？(單位是毫米喔！)"))
    dx = x/30.3
    dy = y/30.3
    dz = z/30.3
    price = float(input("報價？(新台幣)"))

    totalOriginalVolumn = x*y*z
    totalConvertedVolumn = dx*dy*dz
    totalCuft = (dx*dy*dz)/100.0
    totalPrice = totalCuft*price
    print("----------------------------------------------------------")
    print(f"總體機為{totalConvertedVolumn:.4f}立方寸，{totalOriginalVolumn:.4f}立方毫米，{totalCuft:.4f}才")
    print(f"總價為{totalPrice:.4f}新台幣")
```
